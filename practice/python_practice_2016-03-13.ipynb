```python
#第一頁全部內容抓取
import requests 
from bs4 import BeautifulSoup as BS
url = 'http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance'
playload = {  # form data
'method':'search',
'searchMethod':'true',
'searchTarget':'ATM',
'hid_1':'1',
'tenderStatus':'4,5,21,29',
'awardAnnounceStartDate':'105/03/06',
'awardAnnounceEndDate':'105/03/15',
'proctrgCate':'3',
'radProctrgCate':'3',
'hid_2':'1',
'hid_3':'1',
'btnQuery':'查詢'
}
res = requests.post(url, data = playload)
soup = BS(res.text)
print soup
```

### 步驟二_第一頁各自的決標案 URL-case1

```python
div = soup.select('#print_area')[0]
trs = div.select('tr')[1:-1]
for tr in trs:
    print 'http://web.pcc.gov.tw/tps/' + tr.select('a')[0]['href'][3:]
```

### 步驟三_ 抓此頁有多少的標案-case2

```python
from urlparse import urljoin
div = soup.select('#print_area')[0]
trs = div.select('tr')[1:-1]
for tr in trs:
    print urljoin(url,tr.select('a')[0]['href'])
```

### 步驟四_ 計算此頁有多少個分頁數

```python
from math import ceil
num = soup.select('.T11b')[0].text
page_num = int(ceil(float(num)/100))
print page_num 
```

### 步驟四_列出每一分頁的URL

```python
urlformat = 'http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance&searchTarget=ATM&method=search&isSpdt=&pageIndex={}'
for page in xrange(page_num):
    print urlformat.format(page + 1)
```
