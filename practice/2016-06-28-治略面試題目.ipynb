```python
def figure_out(one_file,y):
    count=y
    for neighbor in one_file:   #  section 各自有 2 個小孩  →  sectionName ,articles
        for section in neighbor.getchildren():
            #print (section.getchildren())
            for articles in section.getchildren():
                #print (len(articles.getchildren()))
                lenInElement = len(articles.getchildren())
                Elements=articles.getchildren()
                #print ((articles.getchildren()))
                if lenInElement >0:
                    #print ("YA")
                    for  Ele in Elements:
                        #print ("title:" , Ele.findtext('title'))
                        title=Ele.findtext('title')
                        count=count+1
                        #print ((Ele).iter('authors'))

                        for author in (Ele).iter('authors'):
                            #print (len(author))
                            #print (author.findtext('author'))   #每個文章的第一個作者
                            #print (author.findall('author')[1].text)
                            #print (author.findall('author'))
                            authors=author.findall('author')
                            authors_01=[]
                            for i in authors:
                                authors_01.append(i.text)
                                #print (i.text)
                            #print (authors_01)


                        book_dic={

                            "title":title,
                            "authors":authors_01
                        }
                        #print (book_dic)

                        all_bookcase.append(book_dic)


                        #print ("--------------------------")

    return (all_bookcase,count)
```
