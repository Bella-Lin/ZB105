#### 詳細滿意度列表(約一萬篇留言)

```python
##Lu Lu 超級商城
import requests 
from bs4 import BeautifulSoup as BS
import pandas as pd
from math import ceil
url='https://tw.user.mall.yahoo.com/rating/list?sid=lulus&s=&b=1'
res=requests.get(url)
soup=BS(res.text)
Domain='https://tw.user.mall.yahoo.com/rating/list?sid=lulus&s=&b=%d'
totalPages=soup.findAll('div',{"class":"thd"})[0].select('span')[0]
Pages=totalPages.text
#Page1=Pages.encode('big5')
TotalPage=Pages.split(" ")[1]
#print TotalPage
page_num = int(ceil(float(TotalPage)/10)+1)
#print page_num
ary = []
for i in xrange(0,1):#可以將 page_num 設為欲覽的頁數即可!(ex,1 or 20 or 30)
    #print i+1
    #print Domain%(i+1)
    #url="https://tw.user.mall.yahoo.com/rating/list?sid=lulus&s=&b=1"
    url=Domain%(i+1)
    res=requests.get(url)
    soup=BS(res.text)
    div=soup.findAll("div",{'class': 'bpgbd'})[1]
    ul2=soup.findAll("ul",{'class': 'reply'})
    lis=div.select('li')
    liComment=div.select('lu .reply')
    trs=div.select('tr')
    for tr in trs[1:]:
        td1=tr.select('td')[0]
        reviews=td1.select('em ')[0]#list 指向自己就可以用text
        #print reviews.text
        buyers=tr.select('td')[2]
        #print buyers
        times=tr.select('th')[0]
        #print times
        lisOrder=tr.select('ul')[0]
        lisOrder.encoding='utf-8'
        #print lisOrder.text# orders
        lisComment=tr.select('dd')[0]
        lisComment.encoding='utf-8'
        #print lisComment #orders
        ary.append({
                'review':reviews.text,\
                'buyer':buyers.text,\
                'times':times.text,\
              'comment':lisComment.text,\
                'order':lisOrder.text
              })
df=pd.DataFrame(ary)
df
```
